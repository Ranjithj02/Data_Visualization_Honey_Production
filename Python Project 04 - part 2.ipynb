3. Display the percentage distribution of the data in each year using the pie chart. (2 marks) 

```python
import plotly.express as px
plt.pie(stock.year.value_counts(), autopct = '%.1f%%', radius = 1.0, labels = ['1998', '1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012'])
# display the plot
plt.title('Distribution of data in year')

plt.show()
```

4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings. (2 marks)

5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings. (2 marks) 

```python
sns.scatterplot(x=stock["numcol"], y=stock["prodvalue"], data=stock)
plt.show()
```

         In above scatter plot is strong correlation values

6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings. (2 marks) 

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=stock["year"], y=stock["prodvalue"],data = stock)
plt.show()
```

        In above boxplot compare the prodvalue vs year. In this 2010 year more amount of outlier occurs. 
        low amount of outlier in 1999

7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year') (2 marks) 

```python
sns.pairplot(stock, hue="year", vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue','year'])
plt.show()
```

    In the above pairplot , columns values are  'numcol', 'yield percol', 'total prod', 'prodvalue','year'.
    In that plot  Dark colour values are High and light colour values are Low
    And data are positive correlations

8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue') (2 marks)

```python
correlation = stock[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr ()  
correlation.style.background_gradient (cmap = 'BrBG') 
```

```python
correlation = stock[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr()  
sns.heatmap(correlation,annot=True)


plt.show()
```
